```
#HELPER FUNCTIONS
import numpy as np
class Node:
  def __init__(self, elem = None, weight = 0, next = None):
    self.elem = elem
    self.weight = weight
    self.next = next
#IDEA TAKEN FROM AIB
def printArray(arr):
  for n in arr:
    print(f"Vertex {n.elem}: ",end="")
    n = n.next
    while n!=None:
        print(f"{n.elem},{n.weight}", end=' --> ')
        n = n.next
    print()

def LinkedList(matrix):
  arr = np.array([None]*len(matrix), dtype=Node)
  r, c = matrix.shape
  for i in range(r):
    arr[i] = Node(i)
  for i in range(r):
    for j in range(c):
      if matrix[i][j] != 0:
        head = arr[i]
        while head.next != None:
            head = head.next
        head.next = Node(j, matrix[i][j])
  return arr
```
